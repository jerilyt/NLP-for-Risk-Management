### Encode GBK

```python
import re
def killAnUnseen(s):
    try:
        s.encode('gbk')
        # print("return" + s)
        return s
    except UnicodeEncodeError as err:
        mode = re.findall(r'position ([0-9]*): illegal multibyte sequence', str(err))
        # if mode:
        #     print("position:" + mode[0])
        # print(s[int(mode[0])])
        news = s.replace(s[int(mode[0])], "", 1)
        return killAnUnseen(news)
```

```python
text_data_ = test
clean_content = []
for row_idx in range(text_data_.shape[0]):
    line_content = text_data_.loc[row_idx,'content']
    try:
        line_content.encode('gbk')
    except UnicodeEncodeError:
        line_content  = killAnUnseen(line_content)
    clean_content.append(line_content)
```

```python
def json_to_dataframe(sentimentClassify_data):
    positive_prob_list = []
    confidence_list = []
    negative_prob_list = []
    sentiment_list = []
    
    for row_idx in range(sentimentClassify_data.shape[0]):
        row_result = str(sentimentClassify_data.iloc[row_idx].result)
        positive_prob_list.append(float(re.findall(r'-?\d+\.?\d*e?-?\d*?', row_result)[-4]))
        confidence_list.append(re.findall(r'-?\d+\.?\d*e?-?\d*?', row_result)[-3])
        negative_prob_list.append(float(re.findall(r'-?\d+\.?\d*e?-?\d*?', row_result)[-2]))
        sentiment_list.append(float(re.findall(r'-?\d+\.?\d*e?-?\d*?', row_result)[-1]))
        
    result_data = sentimentClassify_data.copy()
    result_data['positive_prob'] = positive_prob_list
    result_data['confidence'] = confidence_list
    result_data['negative_prob'] = negative_prob_list
    result_data['sentiment'] = sentiment_list
    return result_data
```
